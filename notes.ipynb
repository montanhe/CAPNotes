### About the exam

The CAP exam aims to test applicants on the seven domains needed for analytics professionals, namely:
1. Business Problem Framing
1. Analytics Problem Framing
1. Data
1. Methodology Selection
1. Model Building
1. Deployment
1. Lifecycle management

The exam is based on the Job Task Analysis (JTA). The JTA is a comprehensive description of the duties and responsibilities of a profession, occupation, or specialty area.
